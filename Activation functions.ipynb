```python
import numpy as np
import matplotlib.pyplot as plt
```

### Activation functions

```python
x = np.arange(-10, 11, 0.1)
x = np.delete(x, 10) # we exclude 0 becease of division problem
W0 = 0.5
b0 = 2
```

#### ReLu 

```python
psi = np.dot(x, W0) + b0
y = np.maximum(psi, 0)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y');
```

#### Sigmoid

```python
y = 1/(1+np.exp(-x))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y');
```

#### Tangens activation

```python
y = (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y');
```

#### Step activation

```python
y = [1 if i > 0 else 0 for i in x]
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y');
```

#### Arctangent 

```python
y = np.arctan(x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y');
```

#### Gaussian

```python
y = np.exp(-(np.power(x,2)))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y');
```
