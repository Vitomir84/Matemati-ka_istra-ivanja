### Activation functions

In artificial neural networks, the activation function of a node defines the output of that node given an input or set of inputs. A standard integrated circuit can be seen as a digital network of activation functions that can be "ON" (1) or "OFF" (0), depending on input. This is similar to the linear perceptron in neural networks. However, only nonlinear activation functions allow such networks to compute nontrivial problems using only a small number of nodes, and such activation functions are called nonlinearities.

This script will plot the most known activation functions (ReLu, Sigmoid, Tangens, Step, Arctan, Gaussian) and show how they work in Python. For some activation function, two different inputs will be used. 

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-10, 11, 0.1)
x = np.delete(x, 10) # we exclude 0 becease of division problem
W0 = 0.5
b0 = 2
```

#### ReLu 

```python
psi = np.dot(x, W0) + b0 # with parameters that are learned by some approximation function
y = np.maximum(psi, 0)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y');
```

#### Sigmoid

```python
y = 1/(1+np.exp(-(b0+W0*x)))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y');
```

#### Tangens activation

```python
psi = np.dot(x, W0) + b0
y = (np.exp(psi)-np.exp(-psi))/(np.exp(psi)+np.exp(-psi))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y');
```

#### Step activation

```python
y = [1 if i > 0 else 0 for i in psi]
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y');
```

#### Arctangent 

```python
y = np.arctan(x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y');
```

```python
# with parameters
psi = np.dot(x, W0) + b0
y = np.arctan(psi)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y');
```

#### Gaussian

```python
y = np.exp(-(np.power(x,2)))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y');
```

```python
y = np.exp(-(np.power(psi,2)))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y');
```
