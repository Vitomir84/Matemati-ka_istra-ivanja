With this short example we want to illustrate need for optimisation and heruistics. Brute force approach in Traveller Salesman problem results in huge computational time, while simple greedy algorithm gives ubelievable improvement. This is NP-hard problem in operational research field.

```python
import random
from itertools import permutations
alltours = permutations
```

```python
# racuna udaljenost dva grada
def distance_points(first, second):
    return abs(first-second)
```

```python
# racuna kilometrazu u turi (svim posecenim gradovima)

def distance_tour(aTour):
    return sum(distance_points(aTour[i-1], aTour[i])
              for i in range(len(aTour)))
```

```python
aCity=complex
def generate_cities(number_of_cities):
    seed=111;width=500;height=300 
    random.seed((number_of_cities, seed))
    return frozenset(aCity(random.randint(1, width), random.randint(1, height)) for c in range(number_of_cities))
```

```python
gradovi = generate_cities(10)
gradovi
```

```python
puno_gradova=generate_cities(2000)
```

```python
def shortest_tour(tours):
    return min(tours, key=distance_tour)
```

```python
# kreira sve moguce ture i bira najkracu
def brute_force(gradovi):
    return shortest_tour(alltours(gradovi))
    
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

def visualize_tour(tour, style='bo-'):
    if len(tour)>1000:
        plt.figure(figsize=(15,10))
    start=tour[0:1]
    visualize_segment(tour + start, style)
    visualize_segment(start, 'rD')
```

```python
import matplotlib.pyplot as plt
def visualize_segment(segment, style='bo-'):
    plt.plot([X(c) for c in segment], [Y(c) for c in segment], style, clip_on=False)
    plt.axis('scaled')

```

```python
def X(city): 'X axis'; return city.real
def Y(city): 'Y axis'; return city.imag
```

```python
import time
from collections import Counter
def trgovacki_putnik(algoritam, gradovi):
    t0=time.process_time() 
    tour = algoritam(gradovi)
    t1=time.process_time() 
    assert Counter(tour)==Counter(gradovi)
    visualize_tour(tour)
    print('{}: {} cities = tour length {:.0f} (in {:.3f} sec)'.format(name(algoritam), len(tour), distance_tour(tour),t1-t0))

def name(algoritam): return algoritam.__name__.replace('_tsp', '')
```

```python
trgovacki_putnik(brute_force, gradovi)
```

```python
def greedy_algorithm(gradovi, start=None):
    C=start or first(gradovi)
    tour=[C]
    neposeceni = set(gradovi-{C})
    while neposeceni:
        C=najblizi_sused(C, neposeceni)
        tour.append(C)
        neposeceni.remove(C)
    return tour
```

```python
def first(kolekcija): return next(iter(kolekcija))
```

```python
def najblizi_sused(A, gradovi):
    return min(gradovi, key=lambda C: distance_points(C,A))
```

```python
trgovacki_putnik(greedy_algorithm, gradovi)
```

```python
trgovacki_putnik(greedy_algorithm, generate_cities(100))
```

```python
# Here we can see that even for 10 time less number of cities brute force approach demands 9.135/0.001 more time
trgovacki_putnik(brute_force, generate_cities(10))
```
