### Monte Carlo simulacija 
Simulacija radi po principu generisanja slucajnih tačaka u kvadrat stranice r=1.
Na osnovu udaljenosti tačke centra koordinatnog sistema moguće je odrediti da li se tačka nalazi unutar ili van
zamišljene četvrtine kruga. Kako je četvrtina površine kruga $\frac{r^2\pi}{4}$, a površina kvadrata $r^2$, broj $\pi$ se može aproksimirati kao $\frac{k}{m} * 4 $ gde je k broj tačaka unutar četvrtine kruga a m ukupan broj simuliranih nasumičnih tačaka. Sa povećanjem broja nasumično generisanih tačaka moguće je broj $\pi$ izračunati sa većom prezinošću na uštrb vremenske efikasnosti.

```python
import random
def pi_aproximation(broj_tacaka):

    
    broj_tacaka_unutar_kruga = 0
    for i in range(broj_tacaka):
        x, y = random.random(), random.random()
        if ((x**2+y**2)**0.5) <= 1:
            broj_tacaka_unutar_kruga +=1
        
    aproksimacija_pi = (broj_tacaka_unutar_kruga/broj_tacaka) * 4
    
    return aproksimacija_pi
```

```python
import time
for i in [10, 100, 10000, 1000000, 1000000000]:
    start = time.monotonic()
    print('aproksimirano pi:', pi_aproximation(i))
    end = time.monotonic()
    print(f'vreme potrebno za računanje pi na osnovu {i} tačaka:', round(end-start,4), 'sekundi')
    print(40*'==')
```

```python
def pi_aproximation_za_plotovanje(broj_tacaka):
   
    
    duzina = []
    unutar_kruga = []
    van_kruga = []
    
    broj_tacaka_unutar_kruga = 0
    for i in range(broj_tacaka):
        x, y = random.random(), random.random()
        if ((x**2+y**2)**0.5) <= 1:
            unutar_kruga.append((x,y))
            broj_tacaka_unutar_kruga += 1
        else:
            van_kruga.append((x,y))
    
    aproksimacija_pi = (broj_tacaka_unutar_kruga/broj_tacaka) * 4
    
    return aproksimacija_pi, unutar_kruga, van_kruga
```

```python
import matplotlib.pyplot as plt
for i in [1000, 10000, 100000]:
    aproksimacija_pi, unutar_kruga, van_kruga = pi_aproximation_za_plotovanje(i)
    plt.scatter([x[0] for x in unutar_kruga], [y[1] for y in unutar_kruga], label='unutar_kruga', color='r', s=0.3)
    plt.scatter([x[0] for x in van_kruga],[y[1] for y in van_kruga], label='van_kruga', color='b', s=0.3)
    plt.legend(loc='best')
    plt.show()
```
