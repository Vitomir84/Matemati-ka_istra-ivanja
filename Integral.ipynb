### Integral
Integral predstavlja površinu ispod neke funkcije (krive). Napisan je mali program za računanje integrala.

```python
import math
def integral(funkcija, x1, x2, n):
    if x1 >x2:
        raise AssertionError('x1 mora biti manje od x2')
    ukupna_duzina = x2-x1
    duzina_dela = ukupna_duzina/n
    ukupna_povrsina = 0
    for i in range(n):
        x_value = x1+duzina_dela*i
        y_value = eval(funkcija.replace('x', str(x_value)))
        deo_povrsine = y_value*duzina_dela
        ukupna_povrsina = ukupna_povrsina+abs(deo_povrsine)
    return round(ukupna_povrsina, 5)
```

```python
def integral_sa_linijama_za_plotovanje(funkcija, x1, x2, n):
    if x1 >x2:
        raise AssertionError('x1 mora biti manje od x2')
    ukupna_duzina = x2-x1
    duzina_dela = ukupna_duzina/n
    ukupna_povrsina = 0
    horizontalne_linije = []
    vertikalne_linije = []
    for i in range(n):
        x_value = x1+duzina_dela*i
        y_value = eval(funkcija.replace('x', str(x_value)))
        
        
        deo_povrsine = y_value*duzina_dela
        horizontalne_linije.append(y_value)
        vertikalne_linije.append(x_value)
        ukupna_povrsina = ukupna_povrsina+abs(deo_povrsine)
    return round(ukupna_povrsina, 5), horizontalne_linije, vertikalne_linije
```

```python
ukupna_povrsina, horizontalne_linije, vertikalne_linije = integral_sa_linijama_za_plotovanje('math.sin(x)', 
                                                                                             0,
                                                                                             math.pi/2,
                                                                                             10)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi, 201)
y = np.sin(x)
plt.figure(figsize=(6,6))
plt.plot(x,y)
plt.axhline(0, c='g')
plt.fill_between(x[100:-50],y[100:-50]);
```

```python
x = np.linspace(-np.pi, np.pi, 201)
y = np.sin(x)
plt.figure(figsize=(6,6))
plt.plot(x,y)
for x in vertikalne_linije:
    plt.axvline(x, c='g')
for y in horizontalne_linije:
    plt.axhline(y, c = 'g');
```
